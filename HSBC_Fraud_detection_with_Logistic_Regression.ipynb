```python
data=pd.read_csv('/content/credit_customers (1) - credit_customers (1).csv')
```

```python
for column in data.columns:
  encoder = LabelEncoder()
  if data[column].dtype == 'object' or  'pd.Series' or 'str':  # Check if the column is non-numeric
    data[column] = encoder.fit_transform(data[column])
```
